```python
import spacy
from nltk.stem import PorterStemmer
from nltk.tokenize import word_tokenize
import string

# Load the English model
nlp = spacy.load("en_core_web_sm")

# Initialize the Porter Stemmer
stemmer = PorterStemmer()

# Example text
text = "I'm going to the U.S. tomorrow."

# Process the text with spaCy
doc = nlp(text)

# Extract sentences
sentences = [sent.text for sent in doc.sents]

# Extract words, their lemmas, and filtered tokens
original_words = [token.text for token in doc]
lemmatized_words = [token.lemma_ for token in doc]

# Filtered tokens: remove stop words and punctuation
filtered_tokens = [token.text for token in doc if not token.is_stop and not token.is_punct]


# Display results
print("Original Sentences:", sentences)
print("Original Words:", original_words)
print("Lemmatized Words:", lemmatized_words)

```
